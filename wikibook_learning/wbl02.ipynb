#### 컨테이너 타입 변수
- 파이썬의 변수는 실제 데이터 구조를 바라보는 포인터임. 넘파이 배열에서 예상외의 결과를 초래할 수 있음

```python
## 넘파이 배열 정의
x = np.array([1,3,5])

y = x
```

```python
x, y
```

```python
## x의 요소를 변경
x[1] = -3

x, y  ## 주의 할 것!
```

- 따로 변해야 하는 경우, 깊은 복사 copy() 를 사용해야 함

```python
x= np.array([1,3,5,7,9])
y = x.copy()

x, y
```

```python
x[1] = -3

x, y
```

#### 텐서, 넘파이

#### 합성함수 파이썬

- 아래의 식을 함수로 정의
$$ f(x) = 2 * x^2 + 4 $$  

```python
def f(x):
    return (2 * x**2 + 4)
```

```python
## 넘파이 배열로 x를 정의한다
x = np.arange(-2, 2.25, 0.25)
x
```

```python
## f(x) 실행결과를 y에 대입
y = f(x)
y
```

```python
## 함수 그래프화
plt.figure(figsize=(12,8))
plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
## 세 가지 기본 함수의 정의
def f1(x):
    return(x**2)

def f2(x):
    return(x*2)

def f3(x):
    return(x+2)

# 합성 함수 만들기

x1 = f1(x)
x2 = f2(x1)
y = f3(x2)
```

```python
## 합성함수 값 확인
x1, x2, y
```

```python
## 그래프그리기
plt.figure(figsize=(12,8))
plt.plot(x, y)
plt.grid(True)
plt.show()
```
